```python
#This notebook converts the data provided by James Aird to the format that can be read by our model.
import pandas as pd
import numpy as np
```

```python
N_gals = np.array([[7148, 5452, 3364, 2441, 1718, 495],\
                  [5417, 13387, 13786, 9843, 5874, 1263],\
                  [2933, 8494, 9382, 5728, 940],\
                  [1749, 4654, 4012, 745],\
                  [829, 1607, 382],\
                  [420, 464, 177],\
                  [345, 163, 112]])

for i in range(len(N_gals)):
    N_gals[i] = np.array(N_gals[i])

print(len(N_gals))
```

```python
data = np.loadtxt('./pledd_all.dat', comments='#')
z_low, z_high, mass_low, mass_high, logER, p, p_low, p_high, flag = data.transpose()

mass = 0.5 * (mass_low + mass_high)

z_low_uni = np.unique(z_low)
z_high_uni = np.zeros(len(z_low_uni))
mass_low_uni = np.unique(mass_low)
mass_high_uni = np.zeros(len(mass_low_uni))

for i in range(len(z_low_uni)):
    z_high_uni[i] = np.unique(z_high[np.where(z_low == z_low_uni[i])[0]])
    
for i in range(len(mass_low_uni)):
    mass_high_uni[i] = np.unique(mass_high[np.where(mass_low == mass_low_uni[i])[0]])
    


print(mass_low_uni)   
# print(mass_high_uni)

# print(p_high / p)
# print(p / p_low)

for i in range(len(z_low_uni)):
    
    z_mid = 0.5*(z_low_uni[i] + z_high_uni[i])
    filename = ('./Aird_2018_z%.2f.qpdf' % z_mid)
    f = open(filename, 'w')
    f.write('#zlow: %.1f\n#zhigh: %.1f\n#ref: https://zenodo.org/record/1009605#.W85W6hNKifU\n#type: qpdf_eta\n' % (z_low_uni[i], z_high_uni[i]))
    f.close()
    
#     pick out the points with flags of 1 and 0, respectively.
    ind1 = np.where((z_low == z_low_uni[i]) & (z_high == z_high_uni[i]) & (flag == 1))[0]
    ind0 = np.where((z_low == z_low_uni[i]) & (z_high == z_high_uni[i]) & (flag == 0))[0]
    m_low = mass_low[ind1]
#     print(z_low_uni[i], np.unique(m_low))
    
#     masses = np.unique(mass[ind])
#   undersample the data points for each (redshift, m*) pair.
    for m in np.unique(m_low): 
#         print(m, N_gals[i])
        ind_z_m = np.where((z_low == z_low_uni[i]) & (z_high == z_high_uni[i]) & (mass_low == m) & (flag == 1))[0]
#         pick out the number of galaxies in the corresponding redshift-mass bin.
        N_gal = N_gals[i][np.where(np.unique(m_low) == m)[0]][0]
#         print(N_gal)
#         sub_ind = np.where(mass_low[ind1] == m)
#         pick out the needed Eddington ratios and probabilities.
        ERs = logER[ind_z_m]
        probs = p[ind_z_m]
#         print(ERs)
#       The starting point of the selected ERs is the minimum ER.
        ER_selected = [ERs[0]]
#       this index is used to calculate the starting point of next segment of Eddington ratio.
        ind_seg = 0
        
        while True:
#             When in a logER segment of 0.32, there are more than 5 galaxies, we take 0.32 as a step in logER
            if np.sum(probs[ind_seg:ind_seg + 13]) * N_gal >= 5:
                ER_selected.append(ERs[ind_seg + 12])
                ind_seg += 12
#             If not, we extend the step until getting at least five galaxies.
            else:
                ii = 0
                while np.sum(probs[ind_seg:ind_seg + 13 + ii]) * N_gal < 5:
                    ii += 1
                    if ind_seg + 13 + ii >= len(probs):
                        break
#               if extending to the maximum ER value still can't give us 5 galaxies, we use the maximum ER as the last point.
                try:
                    ER_selected.append(ERs[ind_seg + 13 + ii])
                    ind_seg += (13 + ii)
                except IndexError:
                    ER_selected.append(ERs[-1])
                    ind_seg = len(ERs)
            if ind_seg >= len(ERs):
                break
#         Now we deal with the flag == 0 points. For them, we simply use the central ones, and artificially enlarge their lower errorbar,
#         in order to mimic the fact that such points only provide the upper limits, since the host contamination is significant.
        ind_z_m = np.where((z_low == z_low_uni[i]) & (z_high == z_high_uni[i]) & (mass_low == m) & (flag == 0))[0]
        ERs = logER[ind_z_m]
        ER_selected.append(ERs[len(ERs) // 2])
        
        
#         ER_selected = [ERs[4 * i] for i in range(0, 100) if 4 * i < len(ERs)]
        print(ER_selected)
        for j in range(len(ER_selected)):
            ind_select = np.where((z_low == z_low_uni[i]) & (z_high == z_high_uni[i]) & (mass_low == m) & (logER == ER_selected[j]))
            #The factor of 1.65 comes from the fact that the stipulated uncertainties by James represent 90% credit interval.
            print(np.unique(flag[ind_select]))
            if np.unique(flag[ind_select])[0]: 
                data = np.array([mass[ind_select], logER[ind_select], np.log10(p[ind_select]), np.log10(p_high[ind_select] / p[ind_select]) / 1.65, np.log10(p[ind_select] / p_low[ind_select]) / 1.65])
            else:
                data = np.array([mass[ind_select], logER[ind_select], np.log10(p[ind_select]), np.log10(p_high[ind_select] / p[ind_select]) / 1.65, [10] * len(ind_select)])
            df = pd.DataFrame(data=data.transpose(), columns=['mass', 'eta', 'prob', 'err_h', 'err_l'])
            df.to_csv(filename, sep=' ', header=False, index=False, mode='a+')
    
```
