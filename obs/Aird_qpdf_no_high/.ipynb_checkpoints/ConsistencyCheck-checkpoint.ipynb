```python
#This cell contains the code for plotting the 
#best fitting BQLF from Ueda+2014 paper, ***CTN only***.

h = 0.7
# BLF
A = 3.26e-6 * (h / 0.7)**3
Lstar = 45.48
gamma1 = 0.89
gamma2 = 2.13
p1s = 5.33
beta1 = 0.46

p2 = -1.5
p3 = -6.2
Lp = 45.67
zc1s = 1.86
zc2s = 3.0
La1 = 46.20
La2 = 45.67
alpha1 = 0.29
alpha2 = -0.1

# # XLF
# A = 2.91e-6 * (h / 0.7)**3
# Lstar = 43.97
# gamma1 = 0.96
# gamma2 = 2.71
# p1s = 4.78
# beta1 = 0.84

# p2 = -1.5
# p3 = -6.2
# Lp = 44.0
# zc1s = 1.86
# zc2s = 3.0
# La1 = 44.61
# La2 = 44.0
# alpha1 = 0.29
# alpha2 = -0.1

def zc1_calc(Lx):
    if Lx <= La1:
        return zc1s * (10**(Lx - La1))**alpha1
    else:
        return zc1s

def zc2_calc(Lx):
    if Lx <= La2:
        return zc2s * (10**(Lx - La2))**alpha2
    else:
        return zc2s

def p1_calc(Lx):
    return p1s + beta1 * (Lx - Lp)

def Phi_z0(Lx):
    return A * ((10**(Lx - Lstar))**gamma1 + ((10**(Lx - Lstar))**gamma2))**(-1)

def e(z, Lx):
    zc1, zc2 = zc1_calc(Lx), zc2_calc(Lx)
    p1 = p1_calc(Lx)
    if z <= zc1:
        return (1 + z)**p1
    elif zc1 < z <= zc2:
        return (1 + zc1)**p1 * ((1 + z) / (1 + zc1))**p2
    else:
        return (1 + zc1)**p1 * ((1 + zc2) / (1 + zc1))**p2 * ((1 + z) / (1 + zc2))**p3

def Phi_z(Lx, z):
    return Phi_z0(Lx) * e(z, Lx)
    # return Phi_z0(Lx)
```

# Description of Convolution Procedure

The goal of the convolution is to calculate the quasar luminosity function (either X-ray or bolometric QLF) over a luminosity grid. For each luminosity value $L$ on the grid at a given redshift bin and stellar mass $M_{*,i}$ (i.e. the median stellar mass of the i-th stellar mass bin at a certain redshift bin in Paper II) , we firstly calculate the corresponding $\lambda_{\mathrm{sBHAR}, i}=\frac{k\ L}{2.6\times10^{35}\mathrm{erg}\cdot\mathrm{s}^{-1}\times\frac{M_{*, i}}{M_{\odot}}}$, where $k=1$ for bolometric luminosity, and $k=25$ for X-ray ones.   This formula comes from Paper II (Eq. (2)). 

Secondly, we use our fiducial semi-analytic model output (which fits the Ueda+2014 X-ray QLF data) to calculate the stellar mass function $\Phi\left(M_{*, i}\right)$ for each $M_{*, i}$, and calculate the probability distribution function of $\lambda_{\mathrm{sBHAR}, i}$  given the stellar mass and redshift, $P\left(\lambda_{\mathrm{sBHAR}, i}|M_{*, i}, z\right)$ using the spline interpolation of the data given by James (https://zenodo.org/record/1009605).

Finally, the quasar luminosity function is calculated as below:

$$\phi\left(L, z\right)=\sum_{i}\Phi\left(M_{*, i}\right)\times\Delta\log M_* \times P\left(\lambda_{\mathrm{sBHAR}, i}|M_{*, i}, z\right)\ ,$$

where $\Delta\log M_*$ is the stellar mass bin width used in the data given by James, i.e. 0.5 dex. Given the fact that the plots in Paper II only show $\log \lambda_{\mathrm{sBHAR}}$ as large as 1.5, we omit the stellar mass bins resulting in $\log \lambda_{\mathrm{sBHAR}} > 1.5$ for each luminosity $L$ and redshift bin in the summation.

The calculation of 90% confidence intervals is similar, where we use the ***lower and upper bounds*** of $P\left(\lambda_{\mathrm{sBHAR}, i}|M_{*, i}, z\right)$ in the procedure, instead of the median values.
